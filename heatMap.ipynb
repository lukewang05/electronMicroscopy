```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = np.genfromtxt('outputDistances.csv', delimiter = ',')
#data = np.rot90(data, 4)
#data = np.fliplr(data)
med = np.median(data)
for r in range(len(data)):
    for c in range(len(data[r])):
        data[r][c] = data[r][c] - med
```

```python
# Create data
df = pd.DataFrame(data, columns=np.arange(len(data[0])), index=np.arange(len(data)))
fig, ax = plt.subplots(figsize=(10,12)) 
from matplotlib import cm as cm
yeet = sns.heatmap(df, cmap=cm.get_cmap("RdYlBu"),ax=ax, vmin=-1.58, vmax=1.58)
#fig = yeet.get_figure()
#fig.savefig("300x500heatMap.png")
#yeet = sns.heatmap(df, annot=True, fmt=".2f", annot_kws={"size": 5}, ax=ax)
```
